```python
import pandas as pd

# Load your dataset
data = pd.read_csv("student_performance_data.csv")

# Build dictionary with a few rows (like your ram/ragul/krish example)
student = {
    "StudentID": data["StudentID"].head(3).tolist(),
    "Gender": data["Gender"].head(3).tolist(),
    "Age": data["Age"].head(3).tolist(),
    "StudyHoursPerWeek": data["StudyHoursPerWeek"].head(3).tolist(),
    "AttendanceRate": data["AttendanceRate"].head(3).tolist(),
    "GPA": data["GPA"].head(3).tolist(),
    "Major": data["Major"].head(3).tolist(),
    "PartTimeJob": data["PartTimeJob"].head(3).tolist(),
    "ExtraCurricularActivities": data["ExtraCurricularActivities"].head(3).tolist()
}

# Create DataFrame with custom index (stu1, stu2, stu3)
df = pd.DataFrame(student, index=["1", "2", "3"])
print(df)
```
